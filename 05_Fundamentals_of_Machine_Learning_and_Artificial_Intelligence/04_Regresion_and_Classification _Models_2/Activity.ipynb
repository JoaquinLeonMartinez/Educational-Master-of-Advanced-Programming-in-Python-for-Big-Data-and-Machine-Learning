## 4.1.1 Hallar las predicciones del modelo para el siguiente conjunto de test.

```python
# imports
import pandas as pd
import numpy as np
```

```python
# Cargamos el dataset
df = pd.read_csv('l4p1_test.csv')
df.head()
```

```python
# z1 = 1*6.46 + x1*-1.6 + x2*-3.43 (Activacion de B1)
df['z1'] = 1 * 6.46 + df['x1'] * -1.6 + df['x2'] * -3.43
df.head()
```

```python
# Aplicamos la transformacion no lineal: (sigmoide)
# B1 = h = 1/(1+exp(-z1))
df['B1'] = 1/(1+ np.exp(-df['z1']))
df.head()
```

```python
# z2 = 1*19.64 + x1*-1.35 + x2*-7.52 (Activacion de B2)
df['z2'] = 1*19.64 + df['x1']*-1.35 + df['x2']*-7.52
df.head()
```

```python
# B2 = h = 1/(1+exp(-z2))
df['B2'] = 1/(1+ np.exp(-df['z2']))
df.head()
```

```python
# Calculamos la salida, para ello necesitamos zc1, zc2 y zc3
df['zc1'] = 1*-5.47 + df['B1']*7.61 + df['B2']*1.9
df['zc2'] = 1*-3.76 + df['B1']*-8.21 + df['B2']*7.87
df['zc3'] = 1*3.96 + df['B1']*-3.81 + df['B2']*-8.46
df.head()
```

```python
# Ahora nos toca cacular la sigmode, C1, C2 y C3
df['C1'] = 1/(1+ np.exp(-df['zc1']))
df['C2'] = 1/(1+ np.exp(-df['zc2']))
df['C3'] = 1/(1+ np.exp(-df['zc3']))
df.head()
```

```python
# Comparamos los resultados con los originales:
df['ClaseCalculada'] = df[['C1', 'C2', 'C3']].idxmax(axis=1)
df['ClaseCalculada'] = df['ClaseCalculada'].replace({'C1': 1, 'C2': 2, 'C3': 3})
df.head(12)
```

## 4.1.2 Evaluar el rendimiento del clasificador en dicho conjunto.

```python
# Imports
'''
 Para este apartado utilizare algunas de las funciones de sklearn para ahorrar tiempo 
 ya que lo realmente importante de esta leccion era lo anterior (esta parte ya la hicimmos en la otra actividad)
'''
from sklearn.metrics import confusion_matrix 
```

```python
# Sensibilidad
conf_matrix = confusion_matrix(df['clase'], df['ClaseCalculada'])

true_positives = conf_matrix[1, 1]
false_negatives = conf_matrix[1, 0]

sensitivity = true_positives / (true_positives + false_negatives)

print("Sensibilidad: ", sensitivity)
```

```python
# FP Rate
false_positives = conf_matrix[0, 1]
true_negatives = conf_matrix[0, 0]

fpr = false_positives / (false_positives + true_negatives)

print("FP Rate: ", fpr)
```

```python
# Especifidad

specificity = true_negatives / (true_negatives + false_positives)
print("Especifidad: ", specificity)
```

```python
# Precision

precision = true_positives / (true_positives + false_positives)
print("Precision: ", precision)
```

```python
# F-score
f_score = 2 * (precision * specificity) / (precision + specificity)
print("F-score: ", f_score)
```

El Deep Learning es una forma específica de Machine Learning que se basa en redes neuronales profundas para aprender automáticamente características y representaciones complejas de los datos en lugar de focalizarse en algoritmos y modelos para buscar patrones y relaciones en los datos como haciamos en el Machine Learning.

Se ha demmostrado que es mas eficiente para resolver problemas con una gran cantidad de datos como pueden ser el reconocimiento de imagenes, el procesamiento de lenguaje natural y el reconocimiento por voz.

El fundamento del Deep Learning se basa en la idea de emular el funcionamiento del cerebro humano mediante redes neuronales artificiales profundas, es un concepto similar al visto en esta clase pero llevado a un nivel mucho mas complejo. El concepto clave detrás del Deep Learning es que las redes neuronales con más capas tienen la capacidad de capturar características y patrones más profundos y abstractos de los datos. A medida que los datos fluyen a través de las capas, se realizan operaciones de transformación en los mismos (similares a las que hemmos visto anteriormente pero utilizando un gran abanico de diferentes algoritmos) lo que permite que la red aprenda representaciones cada vez más refinadas y significativas, acercandose de esta forma al funcionamiento del cerebro humano.


## Redes neuronales multicapa

Las redes neuronales multicapa consisten en el concepto que hemos estado trabajando durante la lección. La estructura básica de una red neuronal multicapa consiste en una capa de entrada, una o varias capas ocultas y una capa de salida. Cada capa está compuesta por un conjunto de neuronas, y cada neurona está conectada a las neuronas de las capas adyacentes a través de conexiones ponderadas con lo que hemos denominado pesos, que determinaran cuando se activara la neurona con la que estan conectados.

La capa de entrada en este caso recibe la entrada del problema qe se quiere resolver, esta entrada pueden ser varias neuronas y que cada una represente un pixel de una imagen, pueden representar informacion para tomar una determinada decisión, etc. Además, serán las encargadas de transmitir esta información a las neuronas de las capas intermedias a traves de las conexiones ponderadas mencionadas anteriormente.

Como se ha comentado antes, cada capa oculta adicional le da más complejidad al modelo que estamos generando, dependiendo del tipo de problema que queramos resolver no siempre será recomendable añadir demasiadas capas.

Finalmente, la capa de salida constará de una o mas neuronas que pueden representar un valor booleano, una decisión, etc.

## Redes neuronales convolucionales (CNN)

Las redes neuronales convolucionales son un tipo de arquitectura de redes neuronales especializadas en el procesamiento de datos estructurados en forma de matrices, como imágenes. Estas redes se utilizan principalmente en tareas de visión por computadora, como reconocimiento de objetos, detección de características y segmentación de imágenes. Probableente tambien seriían una buena alternativa a tener en cuenta en problemas como el mostrado en clase de enseñar a la red neuronal a jugar al Super Mario. (si recuerdo bien la entrada era la imagen del estado actual del nivel y la salida la decisión que toma)

La característica principal de estas redes es su capacidad para capturar patrones espaciales y características locales de los datos de entrada. Para esto utiliza capas convolucionales que aplican diferentes filtros a la entrada para extraer características relevantes. Estos filtros se desplazan a lo largo de la imagen, calculando la suma ponderada de los píxeles en cada posición y generando un mapa de características.

Adicionalmente, utilizan "pooling" o capas de agrupación con las caracteristicas mas relevantes de la entrada, de esta forma reducen la dimensionalidad del problema y por lo tanto es mas sencillo de procesar. Además, son eficientes en términos computacionales y permite un aprendizaje rápido.


## Redes neuronales recurrentes (RNN)

Las redes neuronales recurrentes son un tipo de arquitectura de redes neuronales diseñadas para procesar secuencias de datos, donde la información del pasado influye en la predicción del futuro. A diferencia de las redes neuronales convolucionales, las RNN tienen conexiones retroalimentadas, lo que les permite recordar y utilizar información anteriormente procesada.

La característica fundamental de las redes neuronales recurrentes es su capacidad para procesar datos secuenciales de longitud variable. Cada neurona en una RNN tiene una "memoria" interna que almacena la información relevante del pasado y la combina con los nuevos datos de entrada. Esta memoria permite a la red aprender patrones a largo plazo en la secuencia y modelar dependencias temporales. Esto las hace especialmente útiles en aplicaciones donde el orden de los datos es importante, como el procesamiento del lenguaje natural, la traducción automática, la generación de texto y el reconocimiento de voz.

Estas redes tambien presentan algunos problemas como el desvanecimiento o la explosión del gradiente, para los cuales existen diversas tecnicas para tratarlos, en las que ahora no haremos hincapie.
