## Pregunta 1

Los “diagramas de cajas” son un elemento importante de los mismos.Encuentra un ejemplo donde poder mostrar estos gráficos con Python (el que quieras) simplemente se pide mostrar esas Gráficos con el ejemplo.

```python
np.random.seed(42)
data = np.random.normal(0, 5, size=(100,))

# Crear gráfico de diagrama de caja
fig = plt.figure()
ax = fig.add_subplot(111)
ax.boxplot(data)

# Personalizar el gráfico
ax.set_title("Diagrama de Caja")
ax.set_xlabel("Variable")
ax.set_ylabel("Valor")

plt.show()
```

## Pregunta 2

¿ Qué son Q1, Q2, Q3 ?

Se utilizan Q1, Q2 y Q3 para referirse a los cuartiles de un conjunto de datos en un diagrama de caja. Los cuartiles dividen un conjunto de datos ordenados en cuatro partes iguales. Comento brevemente cada uno de ellos:

Q1 (Primer cuartil): Es el valor que divide al conjunto de datos en el primer 25%. Es decir, el 25% de los datos son menores o iguales a Q1, y el 75% son mayores o iguales a Q1.

Q2 (Segundo cuartil): Es equivalente a la mediana de los datos. Divide al conjunto de datos en el 50%, es decir, el 50% de los datos son menores o iguales a Q2, y el 50% son mayores o iguales a Q2.

Q3 (Tercer cuartil): Es el valor que divide al conjunto de datos en el primer 75%. Es decir, el 75% de los datos son menores o iguales a Q3, y el 25% son mayores o iguales a Q3.

Estos cuartiles son útiles para analizar la distribución y dispersión de un conjunto de datos. A continuacion muestro un ejemplo utilizando los datos del ejercicio anterior.

```python
# Calcular los cuartiles
q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50)
q3 = np.percentile(data, 75)

# Crear gráfico de diagrama de caja
fig = plt.figure()
ax = fig.add_subplot(111)
ax.boxplot(data)

# Resaltar los cuartiles en el gráfico
ax.axhline(q1, color='r', linestyle='--', linewidth=2, label='Q1')
ax.axhline(q2, color='g', linestyle='--', linewidth=2, label='Q2 (Mediana)')
ax.axhline(q3, color='b', linestyle='--', linewidth=2, label='Q3')

# Personalizar el gráfico
ax.set_title("Diagrama de Caja con Cuartiles")
ax.set_xlabel("Variable")
ax.set_ylabel("Valor")
ax.legend()

plt.show()
```

## Pregunta 3

¿ Qué son los “outliers” ? ¿ son mostrados en estos diagramas de cajas?

Los "outliers" son valores atípicos o extremos en un conjunto de datos que difieren significativamente del resto de los valores. Estos valores están situados muy lejos de la mayoría de los datos y pueden indicar errores de medición, variaciones inusuales o eventos raros.

En los diagramas de caja, los outliers se representan como puntos individuales que están más allá de los límites llamados "bigotes" del diagrama. Los bigotes representan los valores mínimo y máximo dentro de un rango aceptable, mientras que los outliers se muestran como puntos individuales más allá de esos límites.

En Matplotlib, por defecto, los diagramas de caja muestran los outliers como puntos individuales más allá de los bigotes.

A continuacion muestro un ejemplo de como se representarian:

```python
np.random.seed(42)
data = np.random.normal(0, 1, size=(100,))
outlier = 3.5
data_with_outlier = np.append(data, outlier)

# Crear gráfico de diagrama de caja
fig = plt.figure()
ax = fig.add_subplot(111)
ax.boxplot(data_with_outlier)

# Personalizar el gráfico
ax.set_title("Diagrama de Caja con Outlier")
ax.set_xlabel("Variable")
ax.set_ylabel("Valor")

# Mostrar gráfico
plt.show()
```
